```python
from tensorflow.keras.models import model_from_json
from qkeras.utils import _add_supported_quantized_objects; co = {}; _add_supported_quantized_objects(co)

def mnist_model():
    jsons = open('Models/hls4_example/qkeras_mnist_cnn.json','r').read()
    model = model_from_json(jsons, custom_objects=co)
    model.load_weights('Models/hls4_example/qkeras_mnist_cnn_weights.h5')

    print_qstats(model)

    return model

def mnist_qmodel():
    jsons = open('Models/hls4_example/qkeras_mnist_dense.json','r').read()
    model = model_from_json(jsons, custom_objects=co)
    model.load_weights('Models/hls4_example/qkeras_mnist_dense_weights.h5')

    print_qstats(model)

    return model

def get_cnn_qmodel():
    x = x_in = Input((28,28,1), name="layer0")
    x = QConv2D(
        16, (3, 3),
        kernel_quantizer=quantized_bits(6,1,0,alpha=1),
        bias_quantizer=quantized_bits(6,1,0,alpha=1),
        name="q_conv2d")(x)
    x = QActivation("quantized_relu(3,1)", name="q_relu")(x)
    x = MaxPooling2D((2,2), name='maxp')(x)
    x = QConv2D(
        16, (3, 3),
        kernel_quantizer=quantized_bits(6,1,0,alpha=1),
        bias_quantizer=quantized_bits(6,1,0,alpha=1),
        name="q_conv2d_1")(x)
    x = QActivation("quantized_relu(3,1)", name="1_relu_1")(x)
    x = MaxPooling2D((2,2), name='maxp_1')(x)
    #x = Conv2D(36, (3,3),
    #          name='Conv2')(x)
    #x = MaxPool2D((2,2), name='Max2')(x)
    x = Flatten()(x)
    x = QDense(10, name='q_dense',
              kernel_quantizer=quantized_bits(6,0,0,alpha=1),
              #kernel_quantizer="stochastic_ternary()",
              bias_quantizer=quantized_bits(6,0,0,alpha=1))(x)
    x = Activation("softmax", name='softmax')(x)
    
    model = Model(inputs=x_in, outputs=x)
    
    print_qstats(model)
    
    return model
```

```python
def get_train_test_set(nn_type='Dense'):
    (x_train, y_train), (x_test, y_test) = mnist.load_data()
    
    x_train = x_train / 256.0
    x_test = x_test / 256.0
    
    if (nn_type=='Dense'):
        x_train = x_train.reshape(x_train.shape[0], -1)
        x_test = x_test.reshape(x_test.shape[0], -1)
        
    y_train = to_categorical(y_train, 10)
    y_test = to_categorical(y_test, 10)
    
    return (x_train, y_train), (x_test, y_test)
```

```python
(x_train, y_train), (x_test, y_test) = get_train_test_set(nn_type='Conv2d')
```

```python
model = Sequential()
#model.add(QActivation(activation=quantized_relu(2), input_shape=(784,), name='relu0'))
model.add(QDense(20, input_shape=(784,), name='fc1',
                 kernel_quantizer=quantized_bits(2,0,1,alpha=1), bias_quantizer=quantized_bits(2,0,1,alpha=1),
                 kernel_initializer='lecun_uniform', kernel_regularizer=l1(0.0001)))
model.add(BatchNormalization(epsilon=1e-6, momentum=0.999, name='bn1'))
model.add(QActivation(activation=quantized_relu(3,1), name='relu1'))
#model.add(QDense(15, name='fc2',
#                 kernel_quantizer=quantized_bits(2,0,1,alpha=1), bias_quantizer=quantized_bits(2,0,1,alpha=1),
#                 kernel_initializer='lecun_uniform', kernel_regularizer=l1(0.0001)))
##model.add(BatchNormalization(epsilon=1e-6, momentum=0.999, name='bn1'))
#model.add(QActivation(activation=quantized_relu(3,1), name='relu2'))
model.add(QDense(10, name='output',
                 kernel_quantizer=quantized_bits(4,0,1,alpha=1), bias_quantizer=quantized_bits(4,0,1,alpha=1),
                 kernel_initializer='lecun_uniform', kernel_regularizer=l1(0.0001)))
#model.add(BatchNormalization(epsilon=1e-6, momentum=0.999, name='bnout'))
model.add(Activation(activation='softmax', name='softmax'))

print_qstats(model)
```

```python
def get_cnn_model():
    x = x_in = Input((28,28,1), name="layer0")
    x = Conv2D(24, (3,3),
              input_shape=(28,28,1),
              name='Conv1')(x)
    x = MaxPool2D((2,2), name='Max1')(x)
    x = Conv2D(36, (3,3),
              name='Conv2')(x)
    x = MaxPool2D((2,2), name='Max2')(x)
    x = Flatten()(x)
    x = Dense(10, name='output')(x)
    x = Activation("softmax", name='softmax')(x)
    
    model = Model(inputs=x_in, outputs=x)
    
    print_qstats(model)
    
    return model
```

```python
def get_qmodel():
    x = x_in = Input((784,), name="layer0")
    #x = QActivation("quantized_relu(2)", name='relu0')(x)
    x = QDense(14, name='fc1',
               kernel_quantizer=quantized_po2(4,1),
               #kernel_quantizer="stochastic_ternary()",
               bias_quantizer=quantized_bits(2,0,1))(x)
    x = QActivation("quantized_relu(3,1)", name='relu1')(x)
    x = QDense(14, name='fc2',
               kernel_quantizer=quantized_bits(6,1),
               #kernel_quantizer="stochastic_ternary()",
               bias_quantizer=quantized_bits(2,0,1))(x)
    x = QActivation("quantized_relu(3,1)", name='relu2')(x)
    x = QDense(10, name='output',
               kernel_quantizer=quantized_bits(2,0,1, alpha=1),
               bias_quantizer=quantized_bits(3,0,1))(x)
    x = Activation("softmax", name='softmax')(x)
    
    model = Model(inputs=x_in, outputs=x)
    
    print_qstats(model)
    
    return model
```

```python
def get_qmodel():
    x = x_in = Input((784,), name="layer0")
    x = QActivation("quantized_relu(2)")(x)
    x = QDense(14, name='fc1',
               kernel_quantizer=quantized_bits(6,0,alpha=1),
               bias_quantizer=quantized_bits(5,0,alpha=1))(x)
    x = QActivation("quantized_relu(3,1)", name='relu1')(x)
    x = QDense(10, name='output',
               kernel_quantizer=quantized_bits(3,0,alpha=1),
               bias_quantizer=quantized_bits(6,0,alpha=1))(x)
    x = Activation("softmax", name='softmax')(x)
    
    model = Model(inputs=x_in, outputs=x)
    
    print_qstats(model)
    
    return model
```

```python
from tensorflow_model_optimization.python.core.sparsity.keras import prune, pruning_callbacks, pruning_schedule
from tensorflow_model_optimization.sparsity.keras             import strip_pruning

pruning_params = {"pruning_schedule" : pruning_schedule.ConstantSparsity(0.75, begin_step=2000, frequency=100)}
model = prune.prune_low_magnitude(model, **pruning_params)
```

```python
train = False
if train:
    
    adam = Adam(lr=0.0001)
    model.compile(optimizer=adam, loss=['categorical_crossentropy'], metrics=['accuracy'])
    callbacks = all_callbacks(stop_patience = 1000,
                              lr_factor     = 0.5,
                              lr_patience   = 20,
                              lr_epsilon    = 0.000001,
                              lr_cooldown   = 5,
                              lr_minimum    = 0.000001,
                              verbose       = False,
                              outputDir = MODEL_PATH + 'Qkeras_pruned_CNN')
    callbacks.callbacks.append(pruning_callbacks.UpdatePruningStep())
    with tf.device('/cpu:0'):
        model.fit(x_train, y_train, batch_size=1024,
                  epochs=250, validation_split=0.25, shuffle=True,
                  verbose = 1,
                  callbacks = callbacks.callbacks)
    with tf.device('/cpu:0'):
    # Save the model again but with the pruning 'stripped' to use the regular layer types
        model = strip_pruning(model)
        model.save( MODEL_PATH + 'Qkeras_pruned_CNN/KERAS_check_best_model.h5')
else:
    from tensorflow.keras.models import load_model
    from qkeras.utils import _add_supported_quantized_objects
    co = {}
    _add_supported_quantized_objects(co)
    model = load_model( MODEL_PATH + 'Qkeras_pruned_CNN/KERAS_check_best_model.h5', custom_objects=co)
    model.compile()
    model.summary()
```
